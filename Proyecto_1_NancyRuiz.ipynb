# Ejercicio 2

Escribir un programa que reciba un diccionario con los ingresos
mensuales de los clientes de una empresa, y devuelva una serie
con el ingreso mínimo, máximo y media.

```python
#Diccionario con los ingresos mensuales del cliente
ingresos = {'enero': 18000, 
            'febrero': 21000, 
            'marzo': 19200, 
            'abril': 18900, 
            'mayo': 32300, 
            'junio': 19100, 
            'julio': 19200, 
            'agosto': 18350, 
            'septiembre': 16400, 
            'octubre': 21900, 
            'noviembre': 28000, 
            'diciembre': 45000}

ingresos
```

```python
#Programa que recibe diccionario y devuelve el mínimo, máximo y la media
def estadisticos(ingresos):

  ingresos = pd.Series(ingresos)
  
  #Serie que calcula los estadísticos
  estadisticos = pd.Series([ingresos.min(), ingresos.max(), ingresos.mean()], index=['Ingreso Min', 'Ingreso Max', 'Ingerso Medio'])

  #Imprime los estadísticos
  return print(estadisticos)



#Aplicacion de la función
estadisticos(ingresos)
```

# Ejercicio 3
Escribir programa que genere y muestre por pantalla un DataFrame
con los datos de la tabla (imagen en el doc).
A continuación, generar nuevas columna:
1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas&gt;= 30000, imprimir “Meta Superada”,
caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto&gt;Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril

```python
#Creamos las listas con los valores que va a contener el dataframe
mes = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas = [30500, 35600, 28300, 33900, 42500]
gastos = [22000, 23450, 18100, 35700, 32450]


#Creamos dataframe con la información anterior
balance = pd.DataFrame(data = {'Mes': mes , 'Ventas': ventas, 'Gastos': gastos})
balance
```

```python
#Agregar columna Gasto/Ventas*100
balance['Gastos/Ventas'] = balance['Gastos']/balance['Ventas']*100
balance
```

```python
#Agregar "Meta Superada" Si Ventas >= 30000
balance['Metas'] = balance['Ventas'].apply(lambda i: 'Meta Superada' if i >= 30000 else 'NO Superada')
balance
```

```python
#Agregar el sobre gasto si los Gastos son Mayores a las ventas

#POR QUÉ ASÍ NO FUNCIONA??
#balance['Sobre Gasto'] = balance['Gastos'].apply(lambda i: balance['Gastos'] - balance['Ventas'] if i > balance['Ventas'])

balance.loc[balance['Gastos'] > balance['Ventas'], 'Sobre Gasto'] = balance['Gastos'] - balance['Ventas']
balance
```

```python
#total de las ventas de los últimos 3 meses

sells_last3month = balance['Ventas'].tail(3).sum()
sells_last3month
```

```python
#total de gastos de febrero a abril

feb_abr_df = balance[1:4] #Primero recortamos el data frame con los meses que nos importa observar

gastos_feb_abr = feb_abr_df['Gastos'].sum()
gastos_feb_abr
```
